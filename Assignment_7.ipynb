```python
def upd_col_names(df):
    df.columns = df.columns.str.replace(r"\s+", "_", regex=True)
    return df

df = pd.read_csv("covid_19_country_wise_latest.csv")
df = upd_col_names(df)
df.head()
```

```python
# 5. Check Basic Information â†’ Find column names, data types, and missing values.
print("Column Names:\n", df.columns)

print("Data Types:\n", df.dtypes)

print("Missing Values:\n", df.isnull().sum())

print("Basic Information:\n", df.info)
```

```python
# 6. fill the nan values with the mean of the column
df = df.fillna(df.mean(numeric_only=True))
print(df)
```

```python
# 7. Count the number of unique countries in the dataset.
unique_countries = df["Country/Region"].nunique()
print(unique_countries)
```

```python
# 8.Check if there are any duplicate country entries and remove them if needed.
duplicate_countries = df.duplicated(subset=['Country/Region']).sum()
print("Number of duplicate countries: ", duplicate_countries)

df = df.drop_duplicates(subset=['Country/Region'], keep='first')
print(df)
```

```python
# 9. Find the mean, median, and standard deviation of total cases.
Total_Cases = df["New_cases"]
mean_cases = Total_Cases.mean()
median_cases = Total_Cases.median()
std_cases = Total_Cases.std()

print("Mean of Total Cases:", mean_cases)
print("Median of Total Cases:", median_cases)
print("Standard Deviation of Total Cases:", std_cases)
```

```python
# 10. Find out the string in Deaths Column and replace it with the mean of Deaths column.
df["Deaths"] = pd.to_numeric(df["Deaths"], errors="coerce")

mean_deaths = df["Deaths"].mean()

df.fillna({'Deaths': mean_deaths}, inplace=True)

print(df["Deaths"].head())
```

```python
# 11. Change the datatype of column Deaths
df["Deaths"] = pd.to_numeric(df["Deaths"], errors="coerce").astype("int")
print(df["Deaths"].dtype)
```
